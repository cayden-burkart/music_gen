```python
from audiocraft.models import musicgen
from audiocraft.utils.notebook import display_audio
import torch
```

```python
model = musicgen.MusicGen.get_pretrained('large', device='cuda')

model.set_generation_params(duration=8)

model.lm.load_state_dict(torch.load('models/lm_final.pt'))
```

```python
res = model.generate([
    'Penultimate_Focus_DeepWork_Atmospheric_30_120bpm_HighNEL_Nrmlzd2_expanded_VBR5'
],
    progress=True)
display_audio(res, 32000)
```

```python
res = model.generate([
    'Sun_Spots_Conor_Cinematic2_REPLACEREVISE_1.1_Nrmlzd2_VBR5'
],
    progress=True)
display_audio(res, 32000)
```
